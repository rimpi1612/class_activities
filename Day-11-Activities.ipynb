```python
msft_df = pd.read_csv("MSFT.csv")
msft_df.head()
```

```python
msft_df = msft_df.drop(columns=["volume","open","high","low"])
```

```python
msft_df.head()
```

```python
msft_df["date"]=pd.to_datetime(msft_df["date"],infer_datetime_format=True)
msft_df.head()
```

```python
msft_df.set_index("date")
```

```python
msft_df.plot()
```

```python
daily_returns = msft_df["close"].pct_change(1)
```

```python
daily_returns.plot()
```

```python
cumulative_returns = (1 + daily_returns).cumprod()
cumulative_returns.head()
```

```python
cumulative_returns.plot()
```
